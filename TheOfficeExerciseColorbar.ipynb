```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
df_the_office = pd.read_csv("datasets/office_episodes.csv")
df_the_office.head()
#create a new column, guest_markers. Where has_guests has True, it will put 250, otherwise 25
df_the_office["guest_markers"]=np.where(
    df_the_office["has_guests"] == True, 250, 25) 
df_the_office.head()
```

```python
# Here we should try right away to do a scatterplot. 
# x episode number, y viewership in millions. 
```

```python

# set a color array 0 to 100 for the scaled ratings
colors = df_the_office["scaled_ratings"]*100
sizes =df_the_office["guest_markers"] #create sizes from guest_markers column
#scatter, including sizes and colors for colormap
plt.scatter(df_the_office["episode_number"],df_the_office["viewership_mil"], s=sizes, c=colors, marker = "*", cmap="RdYlGn") #c=colors and cmap allow for a colormap to be added to the scatterplot
plt.title("Popularity, Quality, and Guest Appearances on the Office")
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")
plt.colorbar() #add colorbar to the plot

#find Guest Star:
# 1) find max viewed episode
#2) find index and corresponding guest stars from guest star column
# 3) make a list of it
# 4) choose 1 of them with index slice
# 5) store it in the variable and (put it in the graph)

index_max = df_the_office["viewership_mil"].idxmax()
top_star = list(df_the_office["guest_stars"][77].split(","))[2]
plt.text(75, 21, "guest star!" + top_star)



plt.show()
fig = plt.figure()
plt.rcParams['figure.figsize'] = [11, 7]
```
